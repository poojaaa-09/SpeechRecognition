```
from IPython.display import Audio
#Lets us play the audio directly in the IPython notebook
```

```
from gtts import gTTS 
#Importing the required module
```

```
Mytext = "The rising average temperature of Earth's climate system, called global warming, is driving changes in rainfall patterns, extreme weather, arrival of seasons, and more. Collectively, global warming and its effects are known as climate change"
```

```
language = 'en'
#language for converting the text
```

```
output = gTTS(text=Mytext, lang=language, slow=False) 
```

```
output.save("welcome.mp3")
```

```
sound_file = 'welcome.mp3'
```
